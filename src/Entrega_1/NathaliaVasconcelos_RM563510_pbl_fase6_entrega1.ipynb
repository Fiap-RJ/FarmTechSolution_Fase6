Realizando o teste com 8 imgens retiradas da internet em um contexto diferente para verificar a acurácia do modelo

```python
import os
import subprocess

def get_latest_train_run_folder():
    subfolders = [f.path for f in os.scandir('yolov5/runs/train') if f.is_dir() and '.ipynb_checkpoints' not in f.path]
    latest_folder = max(subfolders, key=os.path.getctime, default=None)
    return latest_folder

latest_run = get_latest_train_run_folder()
if latest_run:
    # COMANDO
    result = subprocess.run(f'python yolov5/detect.py --weights {latest_run}/weights/best.pt --img 320 --source /content/drive/MyDrive/IA-2025/teste --data /content/yolov5/data/fase6.yaml', shell=True, capture_output=True, text=True)
    print(result.stdout)
    print(result.stderr)
else:
    print("Não foi possível encontrar a pasta de treinamento mais recente.")
```

Análise de resultados do teste:
as identificações não foram boas, provavelmente pela qualidade das imagens de treinamento e de validadção. Talvez por ter mais imagens de furadeira o modelo acabou detectando mais furadeiras nas fotos, inclusive quando não eram
